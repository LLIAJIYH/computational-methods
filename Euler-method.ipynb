```python
# imports
from pandas import DataFrame
import numpy as np
from math import exp
from functools import lru_cache
```

```python
def f(x, y):
  return 2*x*y

def y(x, C = 1):
  return C*exp(x**2)

def error(yA, yE):
  return abs(yA - yE)

def divideSegment(a, b, N):
  step = (b-a)/N
  parts = []
  x = a
  for i in range(N+1):
    parts.append(x)
    x+=step
  return parts

@lru_cache(maxsize=None)
def Euler(i, y0, X):
  if i == 0:
    return y0
  return Euler(i - 1, y0, X) + (X[i] - X[i-1]) * f(X[i - 1], Euler(i - 1, y0, X))
```

```python
a = 0
b = 1
y0 = 1
N = 32
X = tuple(np.linspace(0, 1, N+1))
yA = []
yE = []
# yE.append(y0)
errors = []

for i in range(N):
  yE.append(Euler(i, y0, X))
  # yE.append( yE[i - 1] + (X[i] - X[i -1]) * f(X[i - 1], yE[i - 1]))
for i in range(N):
  yA.append(y(X[i]))
  errors.append(error(yA[i], yE[i]))

df = DataFrame(list(zip(X, yE, yA, errors)), columns=["x_i", "Эйлер", "Аналитическое", "Ошибка"])
df
```
